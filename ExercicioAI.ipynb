```python
from sklearn.svm import LinearSVC
from sklearn.svm import SVC
```

```python
#features/rotulos(1 sim, 0 nao)
#longo?
#perna curta?
#faz auau?

porco1 = [0, 1, 0]
porco2 = [0, 1, 1]
porco3 = [1, 1, 0]


cachorro1 = [0, 1, 1]
cachorro2 = [1, 0, 1]
cachorro3 = [1, 1, 1]


# 1 => porco, 0 => cachorro ============ nesse caso porco e representado por 1 e cachoro por 0
dados = [porco1, porco2, porco3, cachorro1, cachorro2, cachorro3]
classes = [1,1,1,0,0,0]

model = SVC()
model.fit(dados, classes)

misterio1 = [1,1,1]
misterio2 = [1,1,0]
misterio3 = [0,1,1]

testes = [misterio1, misterio2, misterio3]
resultados = model.predict(testes)
for resultado in resultados:
    if resultado == 1:
        print("Porco")
    else:
        print("Cachorro")
```

```python
import matplotlib.pyplot as plt

# Definindo os dados
porco1 = [0, 1, 0]
porco2 = [0, 1, 1]
porco3 = [1, 1, 0]
cachorro1 = [0, 1, 1]
cachorro2 = [1, 0, 1]
cachorro3 = [1, 1, 1]

# Dividindo os dados em características separadas
longo = [porco1[0], porco2[0], porco3[0], cachorro1[0], cachorro2[0], cachorro3[0]]
perna_curta = [porco1[1], porco2[1], porco3[1], cachorro1[1], cachorro2[1], cachorro3[1]]
faz_auau = [porco1[2], porco2[2], porco3[2], cachorro1[2], cachorro2[2], cachorro3[2]]

# Plotando o gráfico de dispersão
plt.scatter(longo[:3], perna_curta[:3], color='blue', label='Porco')
plt.scatter(longo[3:], perna_curta[3:], color='red', label='Cachorro')

# Adicionando rótulos e título
plt.xlabel('Longo')
plt.ylabel('Perna Curta')
plt.title('Gráfico de Dispersão das Características')
plt.legend()

# Exibindo o gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Definindo os dados e as classes
porco1 = [0, 1, 0]
porco2 = [0, 1, 1]
porco3 = [1, 1, 0]
cachorro1 = [0, 1, 1]
cachorro2 = [1, 0, 1]
cachorro3 = [1, 1, 1]
dados = [porco1, porco2, porco3, cachorro1, cachorro2, cachorro3]
classes = [1, 1, 1, 0, 0, 0]

# Ajustando o modelo
model = SVC()
model.fit(dados, classes)

# Definindo os mistérios
misterio1 = [1, 1, 1]
misterio2 = [1, 1, 0]
misterio3 = [0, 1, 1]
testes = [misterio1, misterio2, misterio3]

# Prevendo os resultados
resultados = model.predict(testes)

# Reduzindo a dimensionalidade para visualização
pca = PCA(n_components=2)
dados_reduzidos = pca.fit_transform(dados)
testes_reduzidos = pca.transform(testes)

# Plotando os pontos de dados
plt.scatter(dados_reduzidos[:, 0], dados_reduzidos[:, 1], c=classes, cmap='coolwarm', label='Dados')
# Plotando os mistérios
plt.scatter(testes_reduzidos[:, 0], testes_reduzidos[:, 1], c=resultados, cmap='coolwarm', marker='x', label='Mistérios')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.legend(['Porco', 'Cachorro', 'Porco (Mistério)', 'Cachorro (Mistério)'])
plt.title('Distribuição dos Dados e Previsões')
plt.show()
```

```python
from sklearn.metrics import accuracy_score

# Definindo as classes verdadeiras dos dados de teste
classes_teste = [1, 1, 0]

# Calculando a acurácia
acuracia = accuracy_score(classes_teste, resultados)
print("Acurácia do modelo:", acuracia)
```
